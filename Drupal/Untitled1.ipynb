```python
import pandas as pd
import numpy as np
```

```python
image_ids = pd.read_csv("node__field_image_identifier.csv")
```

```python
image_ids.head()
```

```python
lengths = [10, 107, 118, 12, 13]

# Extract the filenames
filenames = image_ids['field_image_identifier_value'].astype(str).tolist()

# Split into chunks and join with '|'
chunks = []
start = 0
for length in lengths:
    end = start + length
    chunk = filenames[start:end]
    chunks.append('|'.join(chunk))
    start = end

# Create a new DataFrame
new_df = pd.DataFrame({
    "chunk_id": range(1, len(chunks) + 1),
    "filenames_combined": chunks
})

print(new_df)
```
