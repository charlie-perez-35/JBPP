Goal here is to just write a little script to merge the proofreading excel sheet with the URLs to the IIIF images in FTP, hopefully to make proofreading a bit simpler.

```python
import json
```

```python
url = 'http://fromthepage.com/iiif/32065742/manifest'
response = requests.get(url)
```

```python
canvases = json.loads(response.text)['sequences'][0]['canvases']
jpgs = []
for i in canvases:
    jpgs.append(i['images'][0]['resource']['@id'])

print(' | '.join(jpgs))
```

```python
def get_jpgs_from_identifier(identifier):
    url = f'http://fromthepage.com/iiif/{identifier}/manifest'
    response = requests.get(url)
    # lazy try/except because I don't want to spend a lot of time troubleshooting this
    try:
        canvases = json.loads(response.text)['sequences'][0]['canvases'] # canvases is where the images are stored
        jpgs = []
        for i in canvases:
            jpgs.append(i['images'][0]['resource']['@id']) # each image id gives a unique IIIF URL to the image
        return ' | '.join(jpgs)
    except:
        return None
```

```python
merged['Image URLs'] = merged.apply(lambda row: get_jpgs_from_identifier(row['*FromThePage ID*']) if row['*FromThePage ID*'] else None, axis = 1)
```
