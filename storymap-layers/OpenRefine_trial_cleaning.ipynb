```python
import spacy

# Load a pre-trained English language model
nlp = spacy.load("en_core_web_sm")

text = "The package should be delivered to 123 Main Street, Anytown, CA 90210."

# Process the text with the NLP model
doc = nlp(text)

# Iterate over named entities in the document
for ent in doc.ents:
    # Check if the entity label is a geopolitical entity (GPE) or location
    if ent.label_ == "GPE" or ent.label_ == "LOC":
        print(f"Potential address component: {ent.text} ({ent.label_})")
```

```python
import pandas as pd
import usaddress

data = {'full_address': ['123 Main St, Anytown, CA 90210', 
                         '456 Oak Ave, Otherville, NY 10001']}
df = pd.DataFrame(data)

def parse_address_components(address):
    try:
        parsed_address = usaddress.tag(address)[0]
        return parsed_address
    except usaddress.RepeatedLabelError:
        pass

df_parsed = df['full_address'].apply(parse_address_components).apply(pd.Series)
df = pd.concat([df, df_parsed], axis=1)

print(df)
```

```python
def insert_newline(text: str) -> str:
    pattern = r'([a-z])([A-Z0-9])'
    return re.sub(pattern, r'\1\n\2', text)
```

```python
datelines = []
for idx, row in export.iterrows():
    dateline = row['Dateline']
    dateline = dateline.replace('\xa0', '')
    dateline = insert_newline(dateline)
    dateline = dateline.split('\n')
    datelines.append(dateline)
```

```python
datelines
```

```python
date_pattern = re.compile(r'\b(?:\d{1,2}[/-]\d{1,2}[/-]\d{2,4}|[A-Z][a-z]+ \d{1,2}, \d{4})\b')

def is_date(s: str) -> bool:
    return bool(date_pattern.search(s))

[[k for k in j if not is_date(k)] for j in datelines]
```

```python
datelines = [
    ["September 23, 1969"],
    ["Not a date"],
    ["February 24, 1975", "Some text"]
]

datelines = [
    [j for j in i if not is_date(j)]
    for i in datelines
]

print(datelines)
```
